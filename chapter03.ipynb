## 21. カテゴリ名を含む行を抽出
記事中でカテゴリ名を宣言している行を抽出せよ．

```python
f = open("data/jawiki-country.json", "r")
for i in f:
    df = json.loads(i)
    if df["title"]=="イギリス":
        Britain = df

text = Britain["text"]

#print(text)

line_list = text.split("\n")


category_list = list()
for i in line_list:
    if re.search("Category", i):
        if not "{{" in i :
            category = i
            category_list.append(category)
for i in category_list:
    print(i)
```

## 22. カテゴリ名の抽出
記事のカテゴリ名を（行単位ではなく名前で）抽出せよ．

```python
f = open("data/jawiki-country.json", "r")
for i in f:
    df = json.loads(i)
    if df["title"]=="イギリス":
        Britain = df

text = Britain["text"]

#print(text)

line_list = text.split("\n")


category_list = list()
for i in line_list:
    if re.search("Category", i):
        if not "{{" in i :
            category = i
            category_list.append(category)
for i in category_list:
    category_name = re.sub(r"\[\[Category:", "", i)
    category_name = re.sub(r"\]", "", category_name)
    category_name = re.sub(r"\|.", "", category_name)
    print(category_name)
```

## 23. セクション構造
記事中に含まれるセクション名とそのレベル（例えば”== セクション名 ==”なら1）を表示せよ．

```python
f = open("data/jawiki-country.json", "r")
for i in f:
    df = json.loads(i)
    if df["title"] == "イギリス":
        Britain = df
            
text = Britain["text"]

line_list = text.split("\n")

for i in line_list:
    if re.search("==", i):
        counter = i.count("=")
        i = re.sub(r"=", "", i)
        level = int(counter/2 - 1)
        print("\t"*(level-1),i,level)
```

## 24. ファイル参照の抽出
記事から参照されているメディアファイルをすべて抜き出せ．

```python
f = open("data/jawiki-country.json", "r")
for i in f:
    df = json.loads(i)
    if df["title"] == "イギリス":
        Britain = df
    
text = Britain["text"]
#print(text)
line_list = text.split("\n")
file_list = list()
for i in line_list:
    if re.search("\<\/ref\>", i):
        continue
    if re.search("\.svg", i):
        file_list.append(i)
    if re.search("\.png", i):
        file_list.append(i)
    if re.search("\.ogg", i):
        file_list.append(i)
    if re.search("\.jpg", i):
        file_list.append(i)
    if re.search("\.JPG", i):
        file_list.append(i)

media_file = list()
        
for i in file_list:
    media = re.sub(r"|.* = ", "", i)
    media = re.sub(r"\[\[.*:", "", media)
    media = re.sub(r"\|.*", "", media)
    media = re.sub(r"\]\]\}\}", "", media)
    print(media)
```

## 25. テンプレートの抽出
記事中に含まれる「基礎情報」テンプレートのフィールド名と値を抽出し，辞書オブジェクトとして格納せよ．

```python
f = open("data/jawiki-country.json", "r")
for i in f:
    df = json.loads(i)
    if df["title"] == "イギリス":
        Britain = df

text = Britain["text"]
line_list = text.split("\n")
britain_dict = dict()
info_list = list()
info_frag = False
# 基礎情報の文字がきて注記までを抽出
for line in line_list:
    if re.search("基礎情報", line):
        info_frag = True
    if line[0] != "|":
        continue
    if info_frag:
        # |を除いて追加
        info_list.append(line[1:])
        if re.search("\|注記 = \<references\/\>", line):
            break
            
for i in info_list:
    # 一つ目の=で区切る
    key, value = i.split("=",1)
    re.sub(r"^\s+", "", key)
    re.sub(r"^\s+", "", value)
    re.sub(r"(\s+)$", "", key)
    re.sub(r"(\s+)$", "", value)
    britain_dict[key] = value
    print(key, value)
```

## 26. 強調マークアップの除去
25の処理時に，テンプレートの値からMediaWikiの強調マークアップ（弱い強調，強調，強い強調のすべて）を除去してテキストに変換せよ（参考: マークアップ早見表）．

```python
def remove_markup(value):
    """マークアップの除去
    
    引数:
    value -- 対象の文字列
    
    戻り値:
    マークアップを除去した文字列
    
    """
    sen = re.sub(r"\'{2,5}", "", value)
    
    return sen

for key, value in britain_dict.items():
    britain_dict[key] =  remove_markup(value)
    print(key, remove_markup(value))
```

## 27. 内部リンクの除去
26の処理に加えて，テンプレートの値からMediaWikiの内部リンクマークアップを除去し，テキストに変換せよ（参考: マークアップ早見表）．

```python
def remove_linkmarkup(value):
    """マークアップの除去
    
    引数:
    value -- 対象の文字列
    
    戻り値:
    リンクマークアップを除去した文字列
    
    """
    if re.search(r"\[\[[^\]]*\]\]", value):
        if re.search(r"ファイル", value):
            return value
        sen = re.sub(r"\[\[[^|]*\|", "", value)
        sen = re.sub(r"\[\[|", "", sen)
        sen = re.sub(r"\]\]", "", sen)
        return sen
            
            
    else:
        return value

for key, value in britain_dict.items():
    britain_dict[key] =  remove_linkmarkup(str(value))
    print(key, remove_linkmarkup(str(value)))
```

## 28. MediaWikiマークアップの除去
27の処理に加えて，テンプレートの値からMediaWikiマークアップを可能な限り除去し，国の基本情報を整形せよ．

```python
# < ~ >以下のの情報はいらない
def remove_allmarkup(value):
    """マークアップの除去
    
    引数:
    value -- 対象の文字列
    
    戻り値:
    可能な限りマークアップを除去した文字列
    
    """
    sen = re.sub(r"\<.*>", "", value)
    sen = re.sub(r"\{\{0\}\}", "", sen)
    return sen

for key, value in britain_dict.items():
    britain_dict[key] =  remove_allmarkup(str(value))
    print(key, remove_allmarkup(str(value)))
```

## 29. 国旗画像のURLを取得する
テンプレートの内容を利用し，国旗画像のURLを取得せよ．（ヒント: MediaWiki APIのimageinfoを呼び出して，ファイル参照をURLに変換すればよい）

```python
import requests
```

```python
filename = "Flag_of_the_United_Kingdom.svg"

session = requests.Session()
url = "https://en.wikipedia.org/w/api.php"
params = {
    'action' : 'query',
    'format' : 'json',
    'prop' : 'imageinfo',
    'titles' : 'File:' + filename,
    'iiprop' : 'url'}
r = session.get(url=url, params=params)
data = r.json()
pages = data['query']['pages']
flag_url = pages[list(pages)[0]]['imageinfo'][0]['url']
print(flag_url)
```
