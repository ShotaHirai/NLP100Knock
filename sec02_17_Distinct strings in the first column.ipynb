```python
f = open("popular-names.txt", "r")
# 行単位で分割
line_list = f.readlines()

#print(line_list)
name = set() # {}で定義すると辞書型
#print(type(name))

# 名前を抽出
for i in line_list:
    item = i.split("\t")
    name.add(item[0])

for i in name:
    print(i)
print(len(name))
```
