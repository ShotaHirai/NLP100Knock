```python
import json
import re
```

```python
f = open("jawiki-country.json", "r")
for i in f:
    df = json.loads(i)
    if df["title"] == "イギリス":
        Britain = df

text = Britain["text"]
line_list = text.split("\n")
Britain_dict = dict()

for i in line_list:
    i = re.sub(r"\<.*\>", "", i)
    i = re.sub(r"\{.*\}", "", i)
    i = re.sub(r"\'", "", i)
    # []を取り除く
    i = re.sub(r"\[\[", "", i)
    i = re.sub(r"\]\]", "", i)
    # <ref~>を取り除く
    i = re.sub(r"注記 =", "", i)
    #i = re.sub(r"styl","", i)
    #i = re.sub(r"clas","", i)
    # =で分割するとkeyとvalueに分かれる
    if re.search("=", i):
        counter = i.count("=")
        # 見出しを弾く
        if counter >= 2:
            break
        print(i)
        dic = i.split("=")
        if re.search("|", dic[0]):
            dic[0] = re.sub(r"\|","", dic[0])
        #print(dic)
        # keyの最後に空白が入っているのでそれ以外を辞書に入れる
        Britain_dict[dic[0][:-1]] = dic[1]
Britain_dict
```
