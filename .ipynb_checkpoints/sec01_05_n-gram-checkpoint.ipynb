```python
def n_gram(Str):
    # 単語ごとに区切る
    str_list = Str.split()
    # 文字列の空白を取り除く
    Str = ""
    for i in range(len(str_list)):
        Str = Str + str_list[i]
    print(Str)
    n = len(str_list)
    # 単語のn-gramを作成
    for i in range(1,n):
        print("単語 ",i,"-gram",sep="")
        for l in range(n-i+1):
            print(str_list[l:l+i])
        print()
    # 文字のn-gramを作成
    n = len(Str)
    for i in range(1,n):
        print("文字 ",i,"-gram",sep="")
        for l in range(n-i+1):
            print(Str[l:l+i])
        print()
```

```python
Str = "I am an NLPer"
```

```python
n_gram(Str)
```
