```python
f = open("popular-names.txt", "r")
line_list = f.readlines()

# 連想配列を用いる
name_freq = {}
for i in line_list:
    name = i.split("\t")[0]
    if name in name_freq:
        name_freq[name] += 1
    else:
        name_freq[name] = 1

# lambda式を使用してvalueで降順ソート
name_freq=sorted(name_freq.items(), key=lambda x:x[1], reverse=True)
# lをプリントすれば回数もわかる
for i,l in name_freq:
    print(l,i)
```
